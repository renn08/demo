### gym_pokemon
#### create a gym env

```python
%reset
import gym
from matplotlib import pyplot as plt

env = gym.make(
    "PokemonGold-v0",
     resolution = (128,128),
     statename="in_the_grass",
)

env.reset()

# let machine play the game randomly and let we record manually for
# a few scenes
img = []
for i in range(1000):
    action = env.action_space.sample()
    observation, reward, done, info = env.step(action)
    env.render()
    if i % 200 == 0:
        img.append(env.render(mode="rgb_array"))

# let we display what we record
for i in range(len(img)):
    plt.imshow(img[i])
    plt.show()

env.close()
```

We can see that the character encounter a SENTRET in the grass and fight with it!

### Recorder
#### non-interact mode

```python
%reset
import gym
from PokemonGoldSimulator.utility import Recorder

env = gym.make(
    "PokemonGold-v0",
    resolution=(256, 256),
    statename="in_the_grass",
)
env.reset()

recorder = Recorder()
for i in range(100):
    action = env.action_space.sample()
    observation, reward, done, info = env.step(action)
    recorder.listen(img=observation, action=action, reward=reward, done=done, state=info['state'])
    env.render()
    # print("action {0}: {1}".format(i, action))
env.close()

recorder.generate_video(filename='recorder_non_interact_example', batch=1, folder="recorder_non_interact_example", save_img=True, resolution=(512, 512))
recorder.collect_data(data_file="recorder_non_interact_example.csv")
```

### Interactor
#### Human agent plays the game

```python
%reset
import gym
from PokemonGoldSimulator.utility.interactor import Interactor
from PokemonGoldSimulator.utility.recorder import Recorder

env = gym.make(
    "PokemonGold-v0",
    resolution = (256,256),
    statename="in_the_grass"
)
env.reset()

# test interactor
interactor = Interactor(env, render_mode="human")
img_seq, state_seq, action_seq, reward_seq, done_seq \
    = interactor.command() # this is how I command the character

# call the interact mode of the recorder to listen to a whole sequecne of data
recorder = Recorder(mode='interact')
recorder.listen(img_seq, state_seq, action_seq, reward_seq, done_seq)

# generate video and collect data in csv file
recorder.generate_video(batch=5, folder="human_agent_example", save_img=True, filename="human_agent_example")
recorder.collect_data(data_file="human_agent_example.csv")
env.close()
```

**In this example, Me, as the human agent, plays the game for 759 actions (which is showed in the output), I successfully walked to a strange man to talk with him, look like he's catching pokemons that only appear at night!**

**By calling the interact mode of the recorder, we can again generate the video and images, along with all the state information saved in a csv file**

### PokemonDataset
One can use the PokemonDataset to access all the data saved by the recorder in a pytorch tensor format, see the example below.

Let we take the data we just generated:
./example/human_agent_example
./example/human_agent_example_img
./example/human_agent_example.csv


```python
%reset

from PokemonGoldSimulator.utility.dataset import PokemonDataset

dataset = PokemonDataset(csv_file="example/human_agent_example.csv", img_dir="example/human_agent_example_img")

print(dataset)
```

```python
# Let see all the data
sample = dataset[:]
print(sample)
```

```python
# Let see each of them, the example I take contains 760 steps,
# for each image, they are of (3,256,256) channel * H * W
print("image tensor shape:")
print(sample['image'].shape)

print("action tensor shape:")
print(sample['action'].shape)

print("observation tensor shape:") # as well as the state tensor shape, containing the information of the pokemon and character
print(sample['observation'].shape)

print("reward tensor shape:")
print(sample['reward'].shape)

print("done tensor shape:")
print(sample['done'].shape)
```

We can also just access a slide of data by passing into idx = 3:10

```python
sample = dataset[3:10]
print(sample['action'].shape)
```

And we can also just access one time step by sending into a single idx, like idx = 10

```python
sample = dataset[10]
print(sample['image'].shape)
```

We can then do a batch on our large image tensors

```python
# suppose we want to do a batch, taking out the 100, 300, 456's data
# let we represent our batch as [100, 300, 456]

batch = [100, 300, 456]
batch_sample = []
for i in batch:
    batch_sample.append(dataset[i])
print(batch_sample)
```

### ImageWrapper

#### RandomCropping

```python
%reset
import gym
from PokemonGoldSimulator.gym_pokemon.image_wrapper import ImageRandomCropWrapper
from PokemonGoldSimulator.utility.recorder import Recorder
from PokemonGoldSimulator.utility.interactor import Interactor

env = ImageRandomCropWrapper(
    gym.make(
        "PokemonGold-v0",
         resolution = (256,256),
         statename="fight_in_progress2",
    ), size = (128, 64)
)# here size is W*H, since we are using pytorch inside
env.reset()
recorder = Recorder(mode="non-interact")
for i in range(100):
    action = env.action_space.sample()
    observation, reward, done, info = env.step(action)
    recorder.listen(img=observation, action=action, reward=reward, done=done, state=info['state'])
recorder.generate_video(filename='random_crop_example', batch=1, resolution=(128,64), folder="random_crop_example", save_img=True)
recorder.collect_data("random_crop_example.csv")
env.close()
```

```python
%reset
import gym
from PokemonGoldSimulator.gym_pokemon.image_wrapper import ImageRandomCropWrapper
from PokemonGoldSimulator.utility.recorder import Recorder
from PokemonGoldSimulator.utility.interactor import Interactor
# it can also be used with interactor

env = ImageRandomCropWrapper(
    gym.make(
        "PokemonGold-v0",
         resolution = (256,256),
         statename="fight_in_progress2",
    ), size = (128, 64)
)# here size is W*H, since we are using pytorch inside
env.reset()

recorder = Recorder(mode="interact")

interactor = Interactor(env)
img_seq, state_seq, action_seq, reward_seq, done_seq = interactor.command()

recorder.listen(img_seq, state_seq, action_seq, reward_seq, done_seq)
recorder.generate_video(batch=5, folder="random_crop_interact", save_img=True, filename="random_crop_interact")
recorder.collect_data(data_file="random_crop_interact.csv")
env.close()
```

#### RandomConvolution

```python
%reset
import gym
from PokemonGoldSimulator.gym_pokemon.image_wrapper import ImageRandomConvWrapper
from PokemonGoldSimulator.utility.recorder import Recorder
from PokemonGoldSimulator.utility.interactor import Interactor

# only to change a little bit can be used to pass in determined kernel
env = ImageRandomConvWrapper(
    gym.make(
        "PokemonGold-v0",
         resolution = (128,128),
         statename="fight_in_progress2",
    ), kernel_size=(3, 3), range=(-10, 11)
)
env.reset()
recorder = Recorder(mode="non-interact")
for i in range(100):
    action = env.action_space.sample()
    observation, reward, done, info = env.step(action)
    recorder.listen(img=observation, action=action, reward=reward, done=done, state=info['state'])
    env.render()
recorder.generate_video(filename='random_conv_example', batch=1, resolution=(128,128), folder="random_conv_example", save_img=True)
recorder.collect_data("random_conv_example.csv")
env.close()
```

**You can view these two example again at ./example with the name random_conv_example.**

#### GuassianNoise

```python
%reset
import gym
from PokemonGoldSimulator.gym_pokemon.image_wrapper import ImageGuassianNoiseWrapper
from PokemonGoldSimulator.utility.recorder import Recorder
from PokemonGoldSimulator.utility.interactor import Interactor


env = ImageGuassianNoiseWrapper(
    gym.make(
        "PokemonGold-v0",
         resolution = (128,128),
         statename="fight_in_progress2",
    ), kernel_size=(5, 9), sigma=(0.1, 5)
)
env.reset()

recorder = Recorder(mode="non-interact")
for i in range(100):
    action = env.action_space.sample()
    observation, reward, done, info = env.step(action)
    recorder.listen(img=observation, action=action, reward=reward, done=done, state=info['state'])
    env.render()

recorder.generate_video(filename='guassian_noise_example', batch=1, resolution=(128,128), folder="guassian_noise_example", save_img=True)
recorder.collect_data("guassian_noise_example.csv")
env.close()
```
